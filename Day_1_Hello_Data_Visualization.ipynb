## Meet Seaborn 🐍

Seaborn is a Python library built on top of Matplotlib that:
- Makes charts beautiful by default
- Works well with Pandas DataFrames
- Has built-in datasets for learning


## Step 1: Importing Libraries (Setup Code)

Run the cell below. Do not modify it for now.


```python
# Core libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Makes plots appear inside the notebook
%matplotlib inline

# Seaborn style
sns.set()
```

## Step 2: Your First Dataset

Seaborn comes with built-in datasets so you don't need to download anything.

Let's load one called **"tips"**.


```python
tips = sns.load_dataset("tips")

# Look at first 5 rows
# TODO: write the correct pandas function to view first rows
```

## Step 4: Your First Plot

Try completing this code:


```python
sns.scatterplot(data=tips, x="total_bill" , y= "tip")

plt.title("Total Bill vs Tip")
plt.show()
```

## Practice Tasks

1. Change the plot title to something more descriptive
2. Try swapping x and y axes
3. Change `hue` to another column like `day`
4. Search online for scatterplot parameters and experiment


```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex")

plt.title("Total Bill vs Tip (Coloured)")
plt.show()
```

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip", size = "size")

plt.title("Plot with size")
plt.show()
```

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "day")

plt.title("Relationship Between Total Bill and Tip Amount (by day)")
plt.show()
```

```python
sns.scatterplot(data = tips, x = "tip", y = "total_bill", hue = "sex")

plt.title("Total Bill vs Tip (Swap axes)")
plt.show()
```
