## Why Exploration Comes First

Before making charts, analysts ask:
- What data do I have?
- Which columns are numerical?
- Which columns are categorical?
- Are there missing values?

> Bad visualization often comes from not understanding the data.


## Step 1: Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

## Step 2: Load the Dataset

We will continue using the **tips** dataset.


```python
tips = sns.load_dataset("tips")

tips.head()
```

## Step 3: Understanding Structure of the Dataset

Run this to understand data types and missing values.


```python
tips.info()
```

Questions to think about:
- Which columns are numerical?
- Which are categorical?
- Are there any missing values?


## Step 4: Statistical Summary of Data

This gives count, mean, min, max, etc.


```python
tips.describe()
```

## Step 5: Unique Values in Categorical Columns

Analysts often check categories before plotting.


```python
tips['day'].unique()
```

## Step 6: Simple Visualization After Exploration

Now that you understand the data better, try this:


```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip")
plt.title("Total Bill vs Tip")
plt.show()
```
