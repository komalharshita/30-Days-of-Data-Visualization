## What is a Line Chart?

A line chart is used when:
- The x-axis is continuous (time, sequence, numbers)
- You want to show trends
- You want to show change over time

> Line charts answer the question: "How is something changing?"


## Step 1: Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

## Step 2: Load the Flights Dataset

This dataset contains airline passenger numbers per month across years.


```python
flights = sns.load_dataset("flights")

# TODO: 

flights.head()
```

## Step 3: Inspect Structure

Check column names and data types.


```python
# TODO: 

flights.info()
```

Think:
- year → time variable
- month → sub-time category
- passengers → numerical value to track


## Step 4: Group Data by Year

We want total passengers per year.


```python
# TODO:

yearly_data = flights.groupby("year")["passengers"].sum().reset_index()
```

## Step 5: Create Line Plot

Plot yearly total passengers.


```python
# TODO:

sns.lineplot(data =yearly_data, x= "year", y="passengers")
plt.title("Total Airline passengers per year ")
plt.show()
```

## Step 6: Monthly Trends Using Hue

Plot passengers over years with hue="month" to observe seasonality.


```python
# TODO:
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.title("Monthly Passenger Trends Over Years")
plt.show()
```
